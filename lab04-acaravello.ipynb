```python
class decisionNode:
    
    def __init__(self, att, classification = None):
        self.attribute = att
        self.children = {}
        self.class_label = classification
        
        
    def addChild(self, child, key):
        self.children[key] = child
        
        
    def addClassification(self, x):
        self.class_label = x
        
        
    
        
```

```python
class decisionTree:
    
    def __init__(self, title = ""):
        self.title = title
        self.tree_node = None
        self.results = []
        
        
    def learning(self, examples, features, default): # Returns a decision tree
        
        if len(examples) == 0:
            return decisionNode("", classification = default)
        elif (len(examples[examples["class"] == "p"]) == len(examples)):
            return decisionNode("", classification = "p")
        elif (len(examples[examples["class"] == "e"]) == len(examples)):
            return decisionNode("", classification = "e")
        elif len(features) == 0:
            return decisionNode("", classification = mode(examples))
        else:
            best = decisionTree.chooseAttribute(features, examples) 
            tree = decisionNode(best)
            features.remove(best)
            for i in best.values:
                exs = examples[examples[best.name] == i]
                subtree = self.learning(exs, features, decisionTree.mode(examples))
                tree.addChild(subtree, i) 
            return tree
        
    
    
    def learn(self, train, features): # Starting method for the recursive learning method
        self.tree_node = self.learning(train, features, "")
    
    
    def predict(self, case): # Takes a single examples and uses it to traverse the tree and makes a prediction about
                             # its edibility
        current_node = self.tree_node
        
        while(current_node.children):
            temp_att = current_node.attribute.name
            for i in current_node.children:
                if (case.loc[0][temp_att] == i):
                    current_node = current_node.children[i]
                    break
        
        return current_node.class_label
    
    
    
    def score(self, test): # Uses predict to predict the ediblity of a set of examples
        correct = 0.0
        wrong = 0.0
        total = len(test)
        for i in range(len(test)):
            ex = test[i:i+1].reset_index(drop=True)
            temp_val = self.predict(ex)
            if (temp_val == ex.loc[0]["class"]):
                correct += 1
            else:
                wrong += 1
        
        self.results.append(correct/total)
        return (correct/total)
    
    
    def resetResults(self):
        self.results = []
    

    def mode(exs):
        if (len(exs[exs["class"] == "p"]) > len(exs[exs["class"] == "e"])):
            return "p"
        elif (len(exs[exs["class"] == "p"]) <= len(exs[exs["class"] == "e"])):
            return "e"
    
    
    def chooseAttribute(atts, ex):
        minVal = 1.0
        bestAtt = None
        for i in atts:
            currentVal = i.entropy(ex)
            if (currentVal < minVal):
                bestAtt = i
                minVal = currentVal
        return bestAtt
    
            
    def pre_order(self, node, depth):
        for i in node.children:
            print(" "*4*depth + i + ": ", end = "")
            if(node.children[i].attribute == ""):
                print("["+node.children[i].class_label+"]")
            else:
                print(node.children[i].attribute.name + "?")
            self.pre_order(node.children[i], depth+1)
    
    
    def __str__(self):
        print(self.tree_node.attribute.name + "?")
        self.pre_order(self.tree_node,1)
        return ""
    
    
```

```python
import numpy
import pandas
import math


class feature:
    
    def __init__(self, name, poss_vals):
        self.name = name
        self.values = poss_vals
    
    
    def partition(self, examples):
        split_examples = []
        for i in self.values:
            split_examples.append(examples[examples[self.name] == i])
            
        return split_examples
    
    
    def entropy(self, examples): 
        total = len(examples)
        split = self.partition(examples)
        ent = 0
        for i in split: 
            pos = len(i[i["class"] == "e"])
            neg = len(i[i["class"] == "p"])
            if (pos == 0 or neg == 0):
                pass
            else:
                magnitude = (pos+neg)/(len(examples[examples["class"] == "p"]) + len(examples[examples["class"] == "e"]))
                val1 = pos/(pos+neg)
                val2 = neg/(pos+neg)
                I = -1*val1*math.log2(val1) + -1*val2*math.log2(val2)
                ent += magnitude*I
        return ent
    
```
